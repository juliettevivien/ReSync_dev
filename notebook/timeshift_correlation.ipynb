```python
import scipy
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
```

```python
df = pd.read_excel('C:\\Users\\Juliette\\OneDrive\\Documents\\PhD\\Projects\\Synchronization project\\timeshift_correlation.xlsx')
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
spearman_corr = stats.spearmanr(x,y)

spearman_corr_coeff = spearman_corr.statistic

spearman_corr_pval = spearman_corr.pvalue
```

```python
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
plt.scatter(x,y,color='blue', alpha=0.5,s=20)
plt.title("Scatter Plot of recording duration vs. timeshift\nwith Linear Regression Line and Spearman Correlation Coefficient")
plt.text(0.1, 0.9, f'Spearman Corr = {spearman_corr_coeff:.2f}\n pvalue = {spearman_corr_pval:.4f} ', transform=plt.gca().transAxes, fontsize=12)
plt.xlabel("recording duration (s)")
plt.ylabel("timeshift (ms)")
coefficients = np.polyfit(x, y, 1)
linear_regression = np.poly1d(coefficients)
plt.plot(x, linear_regression(x), color='black',linewidth=1)
plt.savefig('C:\\Users\\Juliette\\OneDrive\\Documents\\PhD\\Figures\\Sync analysis\\ReSync_grouped\\timeshift correlation\\timeshift correlation.pdf',bbox_inches='tight', dpi=1200)
```
